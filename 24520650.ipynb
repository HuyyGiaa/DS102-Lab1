```python
import numpy as np
import pandas as pd
```

```python
class Linear_Regression:
    def __init__(self, theta: np.ndarray):
        self.theta_ = theta

    def rsme(self, y: np.ndarray, y_hat: np.ndarray) -> float:
        delta = y - y_hat
        return np.sqrt((delta**2).mean())

    def fit(self, y: np.ndarray, X: np.ndarray) -> None:
        cov_1 = np.matmul(X.T, X)
        inv_cov_1  = np.linalg.inv(cov_1)
        self.theta_: np.ndarray = inv_cov_1 @ (X.T @ y)

    def predict(self, X: np.ndarray) -> np.ndarray:
        y_predict = X @ self.theta_.T
        return y_predict
```
